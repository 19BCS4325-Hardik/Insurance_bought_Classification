```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('C:\\work\\insurance.csv')
df_2 = pd.read_csv('C:\\work\\insurance_data.csv')
df_1.head()
```

```python
df = df_1.merge(df_2, left_on='age', right_on='age')
```

```python
df.head()
```

```python
df['bought_insurance'].unique()
```

```python
df.shape
```

```python
target = df['bought_insurance']
data = df.drop(['bought_insurance'],axis=1)
```

```python
data
```

```python
data = pd.get_dummies(data, columns = ['sex'])
data.head()
```

```python
data = pd.get_dummies(data, columns = ['smoker'])
data.head()
```

```python
data = pd.get_dummies(data, columns = ['region'])
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data,target,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
sv = SVC()
rfc = RandomForestClassifier()
```

```python
lr.fit(X_train,y_train)
sv.fit(X_train,y_train)
rfc.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
sv.score(X_test,y_test)
```

```python
rfc.score(X_test,y_test)
```

```python
score = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(data,target,test_size=0.25,random_state=i)
    lr = LogisticRegression()
    lr.fit(X_train,y_train)
    score.append(lr.score(X_test,y_test))
```

```python
np.argmax(score)
```
